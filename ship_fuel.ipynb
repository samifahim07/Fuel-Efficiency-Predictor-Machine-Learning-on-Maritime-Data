```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.metrics import  r2_score , mean_absolute_error,mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv("ship_fuel_efficiency.csv")
df.head()
```

# Data Cleaning

```python
df= df.drop(['CO2_emissions'],axis=1)
df = df.drop(['ship_id'],axis=1)
```

```python
df.isna().sum()
```

There is no nan values

```python
df.duplicated().sum()
```

There is no duplicated values

# Data Spliting

```python
x = df.drop(['fuel_consumption'],axis=1)
y = df[['fuel_consumption']]
```

```python
x.head()
```

# Applying Encoding 

```python
le = LabelEncoder()
```

```python
cat_col = df.select_dtypes(include=[object]).columns
```

```python
for i in cat_col:
    x[i] = le.fit_transform(df[i])
```

# Model Implimentation

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=.8,random_state=42)
```

# Linear Regression Apply

```python
lr = LinearRegression()


# Train Model
linear_model = lr.fit(x_train,y_train)
print("Training Score : ",linear_model.score(x_train,y_train))


# Test Model
print("Test Score : ",linear_model.score(x_test,y_test))

# Prediction 

pred = lr.predict(x_test)
```

# Model Evaluation

```python
# MSE 
mse = mean_squared_error(y_test,pred)
print("MSE : ", mse)

mae = mean_absolute_error(y_test,pred)
print("MAE : ", mae)

r2 = r2_score(y_test ,pred )
print("R2_Score : ", r2)
```

# Decision Tree Regressor Apply

```python
dt = DecisionTreeRegressor()
# Model Training
model_dt = dt.fit(x_train,y_train)
print("Training Score : ", model_dt.score(x_train,y_train))
# Mode
print("Test Score : ", model_dt.score(x_test,y_test))

 # prediction
dt_pred =dt.predict(x_test)
```

# Model Evaluation

```python
# MSE 
mse = mean_squared_error(y_test,dt_pred)
print("MSE : ", mse)

mae = mean_absolute_error(y_test,dt_pred)
print("MAE : ", mae)

r2 = r2_score(y_test ,dt_pred )
print("R2_Score : ", r2)
```

# Lasso Regression

```python
ls = Lasso()
```

```python
# training Model 

model_ls= ls.fit(x_train,y_train)
print("Training Score : ", model_ls.score(x_train,y_train))

# Test Model 
print("Test Data : ", model_ls.score(x_test,y_test))

# prediction

ls_pred = ls.predict(x_test)
```

#  Lesso Model Evaluation

```python
# MSE 
mse = mean_squared_error(y_test,ls_pred)
print("MSE : ", mse)

mae = mean_absolute_error(y_test,ls_pred)
print("MAE : ", mae)

r2 = r2_score(y_test ,ls_pred )
print("R2_Score : ", r2)
```

# Ridge 

```python
rd = Ridge()
```

```python
# training Model 

model_rd= rd.fit(x_train,y_train)
print("Training Score : ", model_rd.score(x_train,y_train))

# Test Model 
print("Test Data : ", model_rd.score(x_test,y_test))

# prediction

rd_pred = rd.predict(x_test)
```

# Model Evaluation

```python
# MSE 
mse = mean_squared_error(y_test,rd_pred)
print("MSE : ", mse)

mae = mean_absolute_error(y_test,rd_pred)
print("MAE : ", mae)

r2 = r2_score(y_test ,rd_pred )
print("R2_Score : ", r2)
```
